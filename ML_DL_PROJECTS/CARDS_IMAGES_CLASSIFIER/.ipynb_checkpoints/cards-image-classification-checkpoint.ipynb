### About Dataset

This is a very high quality dataset of playing card images. All images are 224 X 224 X 3 in jpg format. All images in the dataset have been cropped so that only the image of a single card is present and the card occupies well over 50% of the pixels in the image. There are 7624 training images, 265 test images and 265 validation images. The train, test and validation directories are partitioned into 53 sub directories , one for each of the 53 types of cards. The dataset also includes a .csv file which can be used to load the datasets.\
[__Here__](https://www.kaggle.com/datasets/gpiosenka/cards-image-datasetclassification) you can find this.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd

np.object = np.object_
np.bool = np.bool_
np.int=np.int_

from imutils import paths
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.applications.resnet import ResNet50, ResNet101
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.utils import to_categorical

RANDOM_STATE = 12345
```

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
df = pd.read_csv('data/cards.csv')
df.sample(7, random_state=RANDOM_STATE)
```

```python
df.rename(columns={'class index':'class_index'}, inplace=True)
df.info()
```

```python
df_labels = df.groupby(['class_index', 'labels'], as_index=False).agg({'filepaths':'count'})

_, ax = plt.subplots(figsize=[15, 4], facecolor='lightgrey')
df_labels.plot(kind='bar',
                x='class_index',
                fontsize=8, 
                rot=0, 
                legend=False,  
                edgecolor='black',
                color = 'grey',
                alpha=0.8, 
                ax=ax).set_title('Fig.1 Classes distribution')

plt.xlabel('Class index', fontsize=12)
plt.ylabel('Objects Count', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#grab the list of images in our dataset directory, then initilize
# the list of data (i.e., images) and class images
dict_labels = df_labels['labels'].to_dict()
dict_labels_rev = {v:k for k, v in dict_labels.items()}
imagePaths = list(paths.list_images('data/'))

data = {'train':[], 'valid':[], 'test':[]}
labels = {'train':[], 'valid':[], 'test':[]}

for path in imagePaths:
    label = path.split('\\')[1]
    label = dict_labels_rev[label]

    image = cv2.imread(path)
    image = cv2.resize(image, (128, 128))
    
    set_type = path.split('/')[1].split('\\')[0]
    
    data[set_type].append(image)
    labels[set_type].append(label)
```

```python
fig = plt.figure(figsize=(9,8))
for i in range(12):
    rnd = np.random.randint(0, len(data['train']))
    fig.add_subplot(3, 4, i+1)
    plt.imshow(data['train'][rnd])
    clss = labels['train'][rnd]
    plt.text(0, 140, f"class_index:{clss}", fontsize=10, color='black')
    plt.text(0, 150, f"{dict_labels[clss]}", fontsize=10, color='black')
    plt.xticks([])
    plt.yticks([])
    
plt.tight_layout()  
plt.show()
```

```python
# convert the data into a NumPy array, then preprocess it by scaling
# all pixel intensities th the range [0, 1]
for key in data.keys():
    data[key] = np.array(data[key], dtype=float) / 255.0
    
# encode the labels (which are currently strings) as integers and then
# one-hot encode them   
le = LabelEncoder()
le.fit(df['class_index'])
for key in labels.keys():
    labels[key] = le.transform(labels[key])
    labels[key] = to_categorical(labels[key], 53)
```

```python
trainX = data['train']
valX = data['valid']
testX =  data['test']

trainY = labels['train']
valY = labels['valid']
testY = labels['test']

print(trainX.shape, valX.shape, testX.shape)
print(trainY.shape, valY.shape, testY.shape)
```

```python
# applying data augmentation
aug = ImageDataGenerator(
        rotation_range=20,
        zoom_range=0.1,
        width_shift_range=0.15,
        height_shift_range=0.15,
        shear_range=0.15,
        horizontal_flip=True,
        fill_mode="nearest")
```

```python
# initialize the optimaizer and model with ResNet50 backbone
input_shape = trainX[0].shape
opt = SGD(learning_rate=LEARNING_RATE, momentum=0.9)

backbone = ResNet101(input_shape=input_shape,
                weights='imagenet', 
                include_top=False)

model = Sequential()
model.add(backbone)
model.add(GlobalAveragePooling2D())
model.add(Dense(53, activation='softmax'))

model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])   

LEARNING_RATE = 0.05
BATCH_SIZE = 64
EPOCHS = 100

# train the network
H = model.fit(x=aug.flow(
    trainX, trainY, batch_size=BATCH_SIZE),
    validation_data=(valX, valY),
    epochs=EPOCHS)
```

```python
# evaluate the network
predictions = model.predict(x=testX, batch_size=BATCH_SIZE)
class_report = classification_report(testY.argmax(axis=1),
                           predictions.argmax(axis=1), target_names=list(map(str,le.classes_)),
                           zero_division=0)
print(class_report)
```

```python
plt.plot(H.history['loss'])
plt.plot(H.history['val_loss'])
plt.ylim([0, 5])
plt.legend(['train', 'test'])
plt.grid()
plt.title('Train/test loss')
plt.tight_layout()
plt.show()
```

```python
plt.plot(H.history['accuracy'])
plt.plot(H.history['val_accuracy'])
plt.tight_layout()
plt.legend(['train', 'test'])
plt.grid()
plt.title('Train/test accuracy')
plt.show()
```

### Conclusion
We end up to finetune the "ResNet101" with aproximetly 0.8 of accuracy on the test set. It's not the best result which can be rich on this dataset. In my opinion we should get more attention to split data on train and test sets cause the classification report shows warnings on some classes, where accuracy was equals to zero, cuase there aren't examples in test set.
